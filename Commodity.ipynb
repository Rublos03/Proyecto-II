```python
import pandas as pd

df = pd.read_csv('commodity_trade_statistics_data.csv', low_memory=False)
```

```python
colnames = df.columns
excel2 = pd.DataFrame(columns = colnames)

excel2 = df[df['year'] >= 2005]
excel2 = excel2[excel2['commodity'] == 'ALL COMMODITIES']
excel2 = excel2[excel2['trade_usd'] >= 35000000000]

excel2
```

```python
paises = list(excel2['country_or_area'].unique())
cant_paises = len(paises)
print(excel2['trade_usd'].mean())
print(excel2['trade_usd'].max())
excel2['trade_usd'].min()
print(cant_paises)
```

```python
excel3 = pd.DataFrame(columns = colnames)
excel3 = df[df['year'] >= 2007]
excel3 = excel3[excel3['country_or_area'].isin(paises)]
excel3
```

```python
excel4 = pd.DataFrame(columns = colnames)
excel4 = excel3[excel3['flow'] == 'Import']

excel5 = pd.DataFrame(columns = colnames)
excel5 = excel3[excel3['flow'] == 'Export']
```
